# Merging the Data

```python
#importing required libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating temperature data frame
```

```python
temperature_df = pd.read_csv('D:/DS Projects/731_Final_Project-main/data/Merged_Folder/avg_temp_1970.csv', index_col=0)
```

```python
temperature_df.head()
```

```python
temperature_df = temperature_df.drop(axis=0, columns = {'Year'})
```

```python
temperature_df = temperature_df.rename(index = str, columns = {'dt': 'Year'})
```

```python
temperature_df.head()
```

```python
#adding co2 emissions data
```

```python
co2emissions_df = pd.read_csv('D:/DS Projects/731_Final_Project-main/data/Merged_Folder/global_co2_emissions_1970.csv')
```

```python
co2emissions_df.head()
```

```python
co2emissions_df = co2emissions_df.drop(co2emissions_df.columns[0], axis = 1)
```

```python
co2emissions_df.head()
```

```python
#Joining co2 emissions data with temperature dataframe
```

```python
df1 = pd.merge (temperature_df, co2emissions_df)
```

```python
df1.head()
```

```python
#Adding Fossil Fuel Consumption Data
```

```python
fossil_fuel_df = pd.read_csv('D:/DS Projects/731_Final_Project-main/data/Merged_Folder/Fossil_Fuels_1970.csv')
```

```python
fossil_fuel_df.head()
```

```python
fossil_fuel_df = fossil_fuel_df.drop(fossil_fuel_df.columns[0], axis = 1)
```

```python
fossil_fuel_df.head()
```

```python
#Joining fossil fuel to the main data frame
```

```python
df2 = pd.merge (df1, fossil_fuel_df)
```

```python
df2.head()
```

```python
#Adding sea level data
```

```python
sealevel_df = pd.read_csv('D:/DS Projects/731_Final_Project-main/data/Merged_Folder/global_sea_level_1970.csv')
```

```python
sealevel_df.head()
```

```python
sealevel_df = sealevel_df.drop(sealevel_df.columns[0], axis = 1)
```

```python
sealevel_df = sealevel_df.rename(index = str, columns = {'year': 'Year'})
```

```python
sealevel_df.head()
```

```python
#Joining sea level to the main data frame
```

```python
df3 = pd.merge (df2, sealevel_df)
```

```python
df3.head()
```

```python
#Adding popoulation data
```

```python
population_df = pd.read_csv('D:/DS Projects/731_Final_Project-main/data/Merged_Folder/global_population_1970.csv')
```

```python
population_df.head()
```

```python
population_df = population_df.drop(population_df.columns[0], axis = 1)
```

```python
population_df.head()
```

```python
#Joining Population data to the main data frame
```

```python
df4 = pd.merge (df3, population_df)
```

```python
df4.head()
```
