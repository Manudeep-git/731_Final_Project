```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/Sea_Level/sea_level1960.csv")
df = df[df['year']>=1970]
df.head()
```

```python
fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(12, 8)

sns.pointplot(x='year', y='GMSL',data=df,color='blue',alpha=0.8,figSize=(20,20))
plt.xticks(rotation= 80,ha="right")
plt.xlabel('year',fontweight= 'bold',fontsize = 'x-large')
plt.ylabel('GMSL',fontweight= 'bold',fontsize = 'x-large')
plt.rcParams["xtick.labelsize"] = 10
plt.title('Global Sea Level Rise Over years',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['year'],df['GMSL'],linewidth=3,label = "Global-Median-Sea-Level(GMSL)")
plt.ylabel("Global-Median-Sea-Level(GMSL)",fontweight= 'bold',fontsize = 'x-large')
plt.xlabel("Year",fontweight= 'bold',fontsize = 'x-large')
plt.xticks (fontweight = 'bold')
plt.yticks (fontweight = 'bold')
plt.title("Sea Level rise over years")
plt.legend()
plt.show()
```
